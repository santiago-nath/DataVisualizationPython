```python
africa_ea =  df_can.loc[africa_ea, years_90] #[(df_can['Continent']=='Asia') & (df_can['Region']=='Southern Asia')]
africa_ea.head()
```

```python
africa_ea = africa_ea.transpose()
```

```python
africa_ea.index = africa_ea.index.map(int) # let's change the index values of Haiti to type integer for plotting

count, bin_edges = np.histogram(africa_ea, 15)

africa_ea.plot(kind ='hist', 
          figsize=(10, 6),
          bins=15,
          alpha=0.35,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Immigration from Africa - Eastern Africa')
plt.ylabel('Number of immigrants')
plt.xlabel('90s')

plt.legend(title='Country',title_fontsize=15,loc='upper right', bbox_to_anchor=(1.4, 1))

plt.show() # need this line to show the updates made to the figure
```

```python
# group countries by continents and apply sum() function 
df_2 = df_can.groupby('Region', axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(df_can.groupby('Region', axis=0)))

df_2.head()
```

```python
df_2.shape
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

df_2['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list 
                            )

# scale the title up by 12% to match pctdistance
plt.title('Immigration to Canada by Region [1980 - 2013]', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_2.index, loc='upper left') 

plt.show()
```
